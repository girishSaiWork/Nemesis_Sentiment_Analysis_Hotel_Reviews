```python
# making a column for entity_extraction prediction 
df_pred['entity_extraction'] = [''] * len(df_pred)

# loop through each row in the 'user_review' column
for index, row in df_pred.iterrows():
    # get the user review string from the current row
    user_review = row['user_review']
    # make a prediction using your model for the current user review
    prediction = nlp(user_review)._.cats
     # append the prediction to the 'entity' column for the current row
    df_pred.at[index, 'entity_extraction'] = prediction
```
